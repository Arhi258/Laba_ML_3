```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
```

```python
def f(x):
    return 1/x
```

```python
x_full = np.linspace(0.5, 6, 400)
x_int = np.linspace(1, 5, 100)

area, error = quad(f, 1, 5)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x_full, f(x_full), 'b', label='f(x) = 1/x')
plt.fill_between(x_int, f(x_int), color='skyblue', alpha=0.5, label=f'Площадь ≈ {area:.3f}')

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

plt.title("Визуализация определенного интеграла f(x) = 1/x")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True, linestyle=':')

plt.show()
```
