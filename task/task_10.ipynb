```python
import matplotlib.pyplot as plt
```

```python
months = ['Янв', 'Фев', 'Март', 'Апр', 'Май', 'Июнь', 'Июль', 'Авг', 'Сен', 'Окт', 'Нояб', 'Дек']
water_level = [2.3, 2.4, 2.6, 2.8, 3.0, 3.1, 3.2, 3.1, 3.0, 2.8, 2.6, 2.4]
```

```python
plt.plot(months, water_level, color='blue', marker='o')

plt.title("Уровень воды в озере в течение года")
plt.xlabel("Месяцы")
plt.ylabel("Уровень воды (м)")
plt.grid(True)

plt.show()
```
