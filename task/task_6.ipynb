```python
import matplotlib.pyplot as plt
```

```python
categories = ["Отлично", "Хорошо", "Удовлетворительно", "Неудовлетворительно"]
counts = [20, 35, 30, 15]
```

```python
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=categories, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'orange', 'salmon'])

plt.title("Процентное распределение оценок студентов")
plt.show()
```
