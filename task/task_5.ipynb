```python
import matplotlib.pyplot as plt
```

```python
categories = ["Отлично", "Хорошо", "Удовлетворит.", "Неудовлетворит."]
counts = [20, 35, 30, 15]
```

```python
plt.bar(categories, counts, color=['green', 'blue', 'orange', 'red'])

plt.title("Распределение оценок студентов")
plt.xlabel("Оценки")
plt.ylabel("Количество студентов")

plt.show()
```
