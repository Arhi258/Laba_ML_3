Задание 1. Напишите код, который строит график функции у = x2 на интервале [-10, 10]. Добавьте заголовок, подписи осей и сетку.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = x**2

plt.plot(x, y)
plt.title("График функции y = x^2") 
plt.xlabel("x")                   
plt.ylabel("y")                    
plt.grid(True) 
plt.show()
```

Задание 2. Постройте три линии на одном графике: у = х (синяя, пунктирная линия), $y = x^2$ (зеленая, штрих-пунктирная линия), $y = x^3$ (красная, сплошная линия). Добавьте легенду и сделайте оси одинакового масштаба.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 100)
plt.figure(figsize=(7, 7))

plt.plot(x, x, color='blue', linestyle='--', label='y = x')
plt.plot(x, x**2, color='green', linestyle='-.', label='y = x^2')
plt.plot(x, x**3, color='red', linestyle='-', label='y = x^3')
plt.axis('equal')

plt.legend()
plt.grid(True)
plt.title("Три линии на одном графике")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

Задание 7. Используя mpl_toolkits.mplot3d , постройте 3D-график функции $z = \sin(\sqrt{x^2 + y^2})$ на сетке значений х, у в диапазоне [-5,5].

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(x, y, z, cmap='viridis')

ax.set_title('3D график функции z = sin(sqrt(x^2 + y^2))')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

Задание 12. В каждой задаче требуется: Построить график подинтегральной функции. Вычислить площадь под кривой на заданном отрезке как значение определенного интеграла. Закрасить область под графиком, чтобы визуализировать интеграл. Найдите площадь под функцией: f(x) = 1/x на интервале [1, 5]. Дополнительные требования: В каждой задаче используйте функцию fill_between() для закрашивания площади под кривой. Добавьте подписи осей и заголовок. При необходимости, отобразите оси X и Y для лучшей визуализации.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad

def f(x):
    return 1/x

x_full = np.linspace(0.5, 6, 400)
x_int = np.linspace(1, 5, 100)

area, error = quad(f, 1, 5)

plt.figure(figsize=(8, 6))
plt.plot(x_full, f(x_full), 'b', label='f(x) = 1/x')
plt.fill_between(x_int, f(x_int), color='skyblue', alpha=0.5, label=f'Площадь ≈ {area:.3f}')

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

plt.title("Визуализация определенного интеграла f(x) = 1/x")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True, linestyle=':')

plt.show()
```

Задание 13. Во всех задачах требуется: Построить трехмерный график функции f(x,у) в заданных пределах. Использовать библиотеку Matplotlib для визуализации. Оформить график: добавить заголовок, подписи осей и цветовую карту (если уместно). Построить косинус: f(x, y) = √(x^2+y^2 ) на интервале x, y Є [-5, 5]. Дополнительные требования: Используйте plot_surface() для плотных поверхностей. Используйте plot_wireframe() для каркасных моделей. Настроите цветовую карту через стар. Добавьте подписи осей и заголовки.

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = np.cos(np.sqrt(x**2 + y**2))

fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(x, y, z, cmap='viridis')
ax1.set_title('Плотная поверхность (plot_surface)')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Z')
fig.colorbar(surf, ax=ax1, shrink=0.5, aspect=10)

ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_wireframe(x, y, z, color='blue', linewidth=0.5)
ax2.set_title('Каркасная модель (plot_wireframe)')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Z')

plt.tight_layout()
plt.show()
```
