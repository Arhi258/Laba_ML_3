```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = y * 500 
```

```python
plt.scatter(x, y, c=x, s=sizes, cmap='viridis', alpha=0.7)
plt.title("Диаграмма рассеяния (Задание 3)")
plt.xlabel("Координата X (Цвет)")
plt.ylabel("Координата Y (Размер)")
plt.colorbar(label='Цветовая шкала X')
plt.show()
```
