```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3, 100)
```

```python
plt.figure(figsize=(7, 7))

plt.plot(x, x, color='blue', linestyle='--', label='y = x')
plt.plot(x, x**2, color='green', linestyle='-.', label='y = x^2')
plt.plot(x, x**3, color='red', linestyle='-', label='y = x^3')
plt.axis('equal')

plt.legend()
plt.grid(True)
plt.title("Три линии на одном графике")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```
